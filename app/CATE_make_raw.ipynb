```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
import json
from collections import OrderedDict
from pprint import pprint
```

```python
df_c_all = pd.DataFrame(columns=('Name','Link','API','Label','C_id', 'Total', 'Page_total'))
```

```python
#카테고리별 상황판 만들기 : 1차 카테고리 / 2차 카테고리 / 3차 카테고리 / 카테고리 URL / 페이지네이션 / 상품URL
req = urllib.request
url_c_all = "https://thrivemarket.com/sitemap"
req_c_all = req.urlopen(url_c_all).read()
bs_c_all = BeautifulSoup(req_c_all, "lxml")
bs_c_all = bs_c_all.find_all('ol')[6]
bs_c_all = bs_c_all.find_all('li')[1:]
```

```python
bs_c_all_name = [ i.get_text() for i in bs_c_all]
bs_c_all_link = [ i.a.get('href') for i in bs_c_all]
del bs_c_all_name[175]
del bs_c_all_link[175]
del bs_c_all_name[184]
del bs_c_all_link[184]
del bs_c_all_name[316]
del bs_c_all_link[316]
del bs_c_all_name[317]
del bs_c_all_link[317]
del bs_c_all_name[345]
del bs_c_all_link[345]
df_c_all['Name'] = bs_c_all_name
df_c_all['Link'] = bs_c_all_link
```

```python
df_c_all.drop(df_c_all.index[349:], inplace=True)
```

```python
Label = []
C_id = []
Total = []
#라벨이랑 카테고리ID 갖고오기
for i in bs_c_all_link:
    link_for_api = i
    req_for_api = req.urlopen(link_for_api).read()
    bs_for_api = BeautifulSoup(req_for_api, "lxml")
    str_for_api = str(bs_for_api.find_all(string=re.compile('(filter":)(.*?)(,"productDetails")')))
    r = re.compile('"applied_filters":(.*?),"filter"')
    label_id = r.search(str_for_api).group()
    label = label_id.split('"')[11]
    _id = label_id.split('"')[7]
    Label.append(label)
    C_id.append(_id)
    #
    r2 = re.compile('"asc"},"total":(.*?),"errors"')
    total_tra = r2.search(str_for_api).group()[:-9]
    total = total_tra.split(':')
    Total.append(total[-1])
    print('ok', bs_c_all_link.index(i),i )
```

```python
Total_349 = Total[:-2]
Label_349 = Label[:-3]
C_id_349 = C_id[:-3]
df_c_all['Total'] = Total_349
df_c_all['Label'] = Label_349
df_c_all['C_id'] = C_id_349
```

```python
Count_page = [ int(i)//120 + 1 for i in Total_349]
Count_page_349 = [ str(k) for k in Count_page]
df_c_all['Page_total'] = Count_page_349
```

```python
# url 만들어주기
url_base = 'https://api.thrivemarket.com/products?filter[categories]='
c_id = '3'
p_num = '2'
api_url = url_base + c_id + '&cur_page=' + p_num + '&page_size=120'

#api url 넣으면 원하는 값들 뽑아주기
req_api_url = requests.get(api_url)
bs_api_url = BeautifulSoup(req_api_url.content, "lxml").find('p')
text_api_url = bs_api_url.text
data = json.loads(text_api_url, object_pairs_hook=OrderedDict)
want_to_get = [ 'manufacturer', 'title', 'short_description', 'review_score', 'review_count', 'msrp', 'price', 'url']

for k in range(len(data['products'])):
    a = [ data['products'][k][i] for i in want_to_get]
    print(a, len(a))
```
